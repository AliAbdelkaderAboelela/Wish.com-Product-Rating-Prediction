```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.model_selection import train_test_split
import re
import math
```

```python
df= pd.read_csv('train_new.csv')
df.head()
```

```python
df2= df.drop(columns=['theme','urgency_text','merchant_title','merchant_info_subtitle','currency_buyer','merchant_name','merchant_id','merchant_profile_picture','crawl_month','tags'])
df2.set_index('id',inplace=True)
df2.head()
```

```python
df2.isnull().sum()
```

we see that there is some columns with missing values

#Normalize numerical columns

```python
#min_max_scaler constructor
min_max_scaler = MinMaxScaler()
#extract integer columns
myCols = df2.select_dtypes(include='int64').columns
#normlize integer columns
df2[myCols] = min_max_scaler.fit_transform(df2[myCols])
#normalize float columns
df2[['price','merchant_rating']] = min_max_scaler.fit_transform(df2[['price','merchant_rating']])
```

**rating column has value 6**

```python
print(df2["rating"].value_counts())
mode = df2[df2["rating"]!="?"]["rating"].mode()[0]
df2['rating'] = df2['rating'].replace(6.0, mode)
print(df2["rating"].value_counts())
```

#size column has too many null and similar values

```python
"""helper functions to clean data"""
def clean_product_size(cell):
    
    #replace null values with "ANOTHER_SIZE"
    if not isinstance(cell,str) and math.isnan(cell):
        return "ANOTHER_SIZE"

    #set some regex patterns    
    pattern = re.compile(r'(\bL\b|(?<=[^MF])L|\bX+S|S\b|\bS\b|\dXS|\bM\b|\d?X+L)')
    match = re.search(pattern,cell)
    #if value doesn't match the pattern don't do anything.
    if match is not None:
        cell = match[0]

    #if value has more than X replace the value the number of X's       
    if cell.count('X') > 1 and (cell.endswith('XS') or cell.endswith('XL')):
        cell = f"{cell.count('X')}X{cell[-1]}"
    return cell  

def replace_less_values(col,value_to_be_replaced):
    #extract values with count 2 or less than 2
    bool_lst = [df2[col].value_counts() <= 2]
    lst = list(df2[col].value_counts().index[tuple(bool_lst)])
    #replace them with value of my choice
    df2[col].replace(dict.fromkeys(lst, value_to_be_replaced),inplace=True)    


    
#Count of product_variation_size_id before the cleaning.
print(df2['product_variation_size_id'].value_counts())
df2['product_variation_size_id'] = df2['product_variation_size_id'].str.upper()
df2.product_variation_size_id = df2.product_variation_size_id.apply(clean_product_size)
replace_less_values('product_variation_size_id','ANOTHER_SIZE')
#Count of product_variation_size_id after the cleaning.
df2.product_variation_size_id.value_counts()     
```

#Convert categorical columns to Numerical

```python
#identify all categorical variables
cat_columns = df2.select_dtypes(['object']).columns

#convert all categorical variables to numeric
df2[cat_columns] = df2[cat_columns].apply(lambda x: pd.factorize(x)[0])
```

#Data cleaning
**1-Product color column**

```python
df2['product_color'].value_counts()
```

```python
#since factorize function turns nan values to -1 we will replace -1 with a value of our own 89
df2['product_color']=df2['product_color'].replace(-1,89)
#replace low frequency values

replace_less_values('product_color',89)

df2['product_color'].value_counts()
```

**2-has_urgency_banner**

```python
df2['has_urgency_banner']=df2['has_urgency_banner'].fillna(0)
df2['has_urgency_banner'].value_counts()
```

**3-Origin country**

```python
df2['origin_country'].value_counts()
```

```python
df2['origin_country']=df2['origin_country'].replace(-1,5)
#replace low frequency values

replace_less_values('origin_country',5)

df2['origin_country'].value_counts()
```

```python
df2.isnull().sum()
```

#Hyper Parameter tuning and model training

```python
X_train=df2.drop(['rating'],axis=1)
y_train=df2['rating']
#First model: Decision tree using grid search to tune hyperparameters
DT_param_1 = {
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'min_samples_split':[2,3,4,5],
    'criterion': ["entropy", "gini"]
}

model_dt_1 = DecisionTreeClassifier()


grid_search = GridSearchCV(estimator=model_dt_1, param_grid= DT_param_1, cv=5, n_jobs=-1, verbose=3)


grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_score_)
```

Prediction and saving file

```python
id=df_test['id']
predicted_test_dt = grid_search.predict(df3)
print(f"best hyperparameters found are {grid_search.best_estimator_}")

predicted_test_dt.astype(float)
prediction=pd.DataFrame()
prediction['id']=id
prediction['rating']=predicted_test_dt

prediction.to_csv('ypred',index=False)
```
